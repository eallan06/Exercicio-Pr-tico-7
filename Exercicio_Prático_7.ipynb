```python
import pandas as pd

# 1. Leitura e Preparação
df = pd.read_csv('vendas_brasil_aula4_kpis.csv')
df['data_venda'] = pd.to_datetime(df['data_venda'])

# 2. Cálculo dos KPIs Globais
receita_total = df['receita'].sum()
lucro_total = df['lucro'].sum()
margem_global = (lucro_total / receita_total) * 100
ticket_medio = receita_total / df['pedido_id'].nunique()

print(f"--- KPIs GLOBAIS ---")
print(f"Receita: R$ {receita_total:,.2f}")
print(f"Lucro: R$ {lucro_total:,.2f}")
print(f"Margem: {margem_global:.2f}%")
print(f"Ticket Médio: R$ {ticket_medio:,.2f}\n")

# 3. Análise por Canal (Dimensão)
analise_canal = df.groupby('canal_venda').agg({
    'receita': 'sum',
    'lucro': 'sum',
    'pedido_id': 'nunique'
}).rename(columns={'pedido_id': 'qtd_pedidos'})

analise_canal['margem_%'] = (analise_canal['lucro'] / analise_canal['receita']) * 100
display(analise_canal.sort_values(by='margem_%', ascending=False))
```
